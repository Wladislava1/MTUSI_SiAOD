### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры m, n, min_limit, max_limit, где m и n указывают размер
матрицы, а min_lim и max_lim - минимальное и максимальное значение для
генерируемого числа . По умолчанию при отсутствии параметров принимать следующие
значения:
m = 50
n = 50
min_limit = -250
max_limit = 1000 + (номер своего варианта)

```python
import random
import time

def make_matrix(rows=50, cols=50, min_val=-250, max_val=1021):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            random_number = random.randint(min_val, max_val)
            row.append(random_number)
        matrix.append(row)
    return matrix
my_matrix = make_matrix(rows=20, cols=20)
for row in my_matrix:
    print(row)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах


```python
import timeit
import random

def selection_sort_matrix(matrix):
    for row in matrix:
        for i in range(len(row)):
            min_index = i
            for j in range(i + 1, len(row)):
                if row[j] < row[min_index]:
                    min_index = j
            row[i], row[min_index] = row[min_index], row[i]
time = timeit.timeit(lambda: selection_sort_matrix(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```

```python
def insertion_sort_row(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
def sort_matrix_insertion(matrix):
    for row in matrix:
        insertion_sort_row(row)
time = timeit.timeit(lambda: sort_matrix_insertion(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```

```python
def bubble_sort_row(row):
    n = len(row)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if row[j] > row[j + 1]:
                row[j], row[j + 1] = row[j + 1], row[j]

def sort_matrix_bubbles(matrix):
    for row in matrix:
        bubble_sort_row(row)
time = timeit.timeit(lambda: sort_matrix_bubbles(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```

```python
def shell_sort_row(row):
    n = len(row)
    gap = n // 2 #=1
    # while gap < n // 3:
        # gap = gap * 3 + 1
    while gap > 0:
        for i in range(gap, n):
            temp = row[i]
            j = i
            while j >= gap and row[j - gap] > temp:
                row[j] = row[j - gap]
                j -= gap
            row[j] = temp
        gap //= 2 #//=3

def sort_matrix_shell(matrix):
    for row in matrix:
        shell_sort_row(row)
time = timeit.timeit(lambda: sort_matrix_shell(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```

```python
def tournament_sort(arr):
    def tournament_helper(start, end):
        if start == end:
            return [arr[start]]
        mid = (start + end) // 2
        left_half = tournament_helper(start, mid)
        right_half = tournament_helper(mid + 1, end)
        return merge(left_half, right_half)

    def merge(left, right):
        result = []
        while left and right:
            if left[0] < right[0]:
                result.append(left.pop(0))
            else:
                result.append(right.pop(0))
        result.extend(left or right)
        return result

    return tournament_helper(0, len(arr) - 1)
def sort_matrix_tournament(matrix):
    for row in matrix:
        tournament_sort(row)
time = timeit.timeit(lambda: sort_matrix_tournament(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [] #<
    middle = [] 
    right = [] #>
    for x in arr:
        if x < pivot:
            left.append(x)
        elif x == pivot:
            middle.append(x)
        else:
            right.append(x)
    return quick_sort(left) + middle + quick_sort(right)
    
def sort_matrix_quick(matrix):
    for i in range(len(matrix)):
        matrix[i] = quick_sort(matrix[i]) 
time = timeit.timeit(lambda: sort_matrix_quick(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```

```python
def heap_sort_row(arr, n, i):
    largest = i #нулевой
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heap_sort_row(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1): #первая куча
        heap_sort_row(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heap_sort_row(arr, i, 0)
def sort_matrix_heap(matrix):
    for i in range(len(matrix)):
       heap_sort(matrix[i])
time = timeit.timeit(lambda: sort_matrix_heap(my_matrix), number=1)
print(f"Время выполнения: {time} секунд")
for row in my_matrix:
     print(row)
```
