### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def insert(self, item):
        self.items.insert(0, item)

    def append(self, item):
        self.items.append(item)

    def remove_front(self):
        if not self.is_empty():
            return self.items.pop(0)

    def remove_rear(self):
        if not self.is_empty():
            return self.items.pop()

    def size(self):
        return len(self.items)
    
    def get_item(self, index):
        if index < len(self.items):
            return self.items[index]
        else:
            return None


def sort_books(input_file, output_file):
    deque1 = Deque()
    deque2 = Deque()

    with open(input_file, "r", encoding="utf-8") as file:
        for line in file:
            book = line.strip()
            deque1.append(book)

    while not deque1.is_empty():
        min_value = deque1.remove_front()
        
        for _ in range(deque1.size()):
            item = deque1.remove_front()
            if item < min_value:
                deque1.append(min_value)
                min_value = item
            else:
                deque1.append(item)

        deque2.append(min_value)

    with open(output_file, "w", encoding="utf-8") as file:
        while not deque2.is_empty():
            file.write(deque2.remove_front() + "\n")

sort_books("books.txt", "sorted_books.txt")

```

Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.

```python
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def insert(self, item):
        self.items.insert(0, item)

    def append(self, item):
        self.items.append(item)

    def remove_front(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None

    def remove_rear(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def size(self):
        return len(self.items)

    def get_front(self):
        return self.items[0] if not self.is_empty() else None


def transcript(deque, encrypted_file, decrypted_file):
    with open(encrypted_file, 'r') as file:
        encrypted_text = file.read()

    decrypted_text = ""

    for char in encrypted_text:
        found = False

        for _ in range(deque.size()):
            current = deque.remove_front()
            deque.append(current)

            if current == char:
                found = True
                break

        if found == True:
            deque.insert(deque.remove_rear())
            deque.insert(deque.remove_rear())
            deque.insert(deque.remove_rear())
            decrypted_text += deque.get_front()

    with open(decrypted_file, 'w') as file:
        file.write(decrypted_text)

    print(f"Расшифрованное сообщение записано в файл {decrypted_file}")

deque = Deque()
symbols = "abcdefghijklmnopqrstuvwxyz"
for symbol in symbols:
    deque.append(symbol)

transcript(deque, "encrypted.txt", "decrypted.txt")
```

Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
class Stack:
    def __init__(self, name):
        self.stack = []
        self.name = name
    
    def push(self, item):
        self.stack.append(item)
    
    def pop(self):
        return self.stack.pop()
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)
    
    def __str__(self):
        return str(self.stack)

with open('disks.txt', 'r') as file:
    disks = list(map(int, file.readlines()))

A = Stack("A")
B = Stack("B")
C = Stack("C")

for disk in sorted(disks, reverse=True):
    A.push(disk)

def pyramid_build(n, source, target, destination): 
    if n == 1:
        disk = source.pop()
        target.push(disk)
        print(f"Перенесли диск {disk} с {source.name} на {target.name}")
        print_stacks()
        return
    pyramid_build(n-1, source, destination, target) 
    
    disk = source.pop()
    target.push(disk)
    print(f"Перенесли диск {disk} с {source.name} на {target.name}")
    print_stacks()
    

    pyramid_build(n-1, destination, target, source)

def print_stacks():
    print(f"{A.name}: {A}")
    print(f"{B.name}: {B}")
    print(f"{C.name}: {C}")
    print("-" * 30)

pyramid_build(A.size(), A, C, B)

   
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0
    
def check_brackets_balance(filename):
    stack = Stack()
    with open(filename, 'r') as file:
        for line in file:
            for char in line:
                if char == '(':
                    stack.push(char)
                elif char == ')':
                    if stack.is_empty():
                        print(f"Нарушен баланс скобок: нет открывающей скобки для '{char}'")
                        return False
                    else:
                        stack.pop()
    if stack.is_empty():
        print("Скобки сбалансированы.")
        return True
    else:
        print(f"Нарушен баланс скобок: остались открывающие скобки.")
        return False

check_brackets_balance("program().txt")
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0


def print_symbols():
    digit_stack = Stack()
    letter_stack = Stack()
    other_stack = Stack()

    with open('symbols.txt', 'r', encoding='utf-8') as file:
        for line in file:
            for char in line:
                if char.isdigit():
                    digit_stack.push(char)
                elif char.isalpha():
                    letter_stack.push(char)
                else:
                    other_stack.push(char)

    print("Цифры:")
    temp_stack = Stack()
    while not digit_stack.is_empty():
        temp_stack.push(digit_stack.pop())
    while not temp_stack.is_empty():
        print(temp_stack.pop(), end='')
    print()

    print("Буквы:")
    temp_stack = Stack()
    while not letter_stack.is_empty():
        temp_stack.push(letter_stack.pop())
    while not temp_stack.is_empty():
        print(temp_stack.pop(), end='')
    print()

    print("Остальные символы:")
    temp_stack = Stack()
    while not other_stack.is_empty():
        temp_stack.push(other_stack.pop())
    while not temp_stack.is_empty():
        print(temp_stack.pop(), end='')
    print()
print_symbols()

```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.


```python
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def append(self, item):
        self.items.append(item)

    def remove_front(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None

    def size(self):
        return len(self.items)


def print_numbers(filename):
    deque = Deque()
    negative_count = 0

    with open(filename, 'r') as file:
        for line in file:
            for num in line.split():
                num = int(num)
                deque.append(num) 
                if num < 0:
                    negative_count += 1

    print("Отрицательные числа:")
    for _ in range(deque.size()):
        num = deque.remove_front()
        if num < 0:
            print(num, end=' ')
        deque.append(num)

    print("\nПоложительные числа:")
    for _ in range(deque.size()):
        num = deque.remove_front()
        if num > 0:
            print(num, end=' ')
        deque.append(num)


print_numbers("numbers.txt")

      
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0


def reverse_lines(input_filename, output_filename):
    line_stack = Stack()

    with open(input_filename, 'r') as input_file:
        for line in input_file:
            line_stack.push(line)

    with open(output_filename, 'w') as output_file:
        while not line_stack.is_empty():
            output_file.write(line_stack.pop())

        print(f"Строки записаны в обратном порядке в файл {output_filename}")

reverse_lines("input.txt", "reverse_output.txt")
```
